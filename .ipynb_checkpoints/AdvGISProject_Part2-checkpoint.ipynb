The following script analyzes the output from the Part 1 script of this project, which is a shapefile of vessel points within Cape Cod Bay. The analysis from part 1 provided a `NEAR_FC` field, that this script uses to examine ships flagged near two different points of interest. After wrangling and cleaning the data, the script visualizes the single voyage to Provincetown port and the three voyages through Cape Cod Canal that were found to be breaking seasonal speed limits.

#### Setup -- import packages, read in data

```python
# import packages
import geopandas as gpd
import pandas as pd
import contextily as cx
```

```python
# point to our near_analysis output from part 1
near_output = r"..\Data\Output\ShipsCCB_Cluster_NearOutput.shp"

# read in attributes as geodataframe
CCBships = gpd.read_file(near_output)
```

```python
# read in vessel data as dataframe
VesselData = pd.read_csv(r"..\Data\Output\VesselJoin.csv")
```

#### Data wrangling -- joining and cleaning

```python
# merge tables using MMSI as join key
FullData = CCBships.merge(VesselData, how = 'left', on = 'MMSI')
FullData.info()
```

```python
# select only fields of interest
CleanData = FullData[['MMSI','VoyageID','VesselType.y','Length','Width','SOG','NEAR_FC','geometry']]
CleanData.info()
```

#### Analysis and Visualization
_Which ships stopped in Provincetown?_

```python
# Subset points that were flagged 100m from Provincetown
PTown = CleanData.query('NEAR_FC == "V:\\CourseProject\\ShipTracker.gdb\\PTown_feature"')
PTown.info()
```

```python
# Identify the voyage and ship ID that went to Provincetown
PTown_voyage_count = PTown['VoyageID'].nunique()
PTown_voyage_ID = PTown['VoyageID'].unique()[0]
PTown_MMSI = PTown['MMSI'].unique()[0]
print(f"There was {PTown_voyage_count} voyage to Provincetown in Jan 2014: Voyage {PTown_voyage_ID}, MMSI {PTown_MMSI}")
```

```python
# subset voyage 35 points
Voyage35 = CCBships.query('VoyageID == 35')
```

```python
# visualize the voyage to and from Provincetown
Voy35 = Voyage35.to_crs(epsg=3857) # reproject points to Web Mercator (to be compatible with basemap display)
Voy35_map = Voy35.plot(figsize=(10,8), alpha=.5, markersize=5) # plot the reprojected points
cx.add_basemap(Voy35_map) # add basemap (Web Mercator projection)
```

_Which ships went through Cape Cod Canal?_

```python
# Subset points that were flagged 100m from the Cape Cod Canal
Canal = CleanData.query('NEAR_FC == "V:\\CourseProject\\ShipTracker.gdb\\CanalNE_feature"')
Canal.info()
```

```python
# Identify the voyages and ships that went through the canal
Canal_voyage_count = Canal['VoyageID'].nunique()
Canal_MMSI_count = Canal['MMSI'].nunique()
print(f"There were {Canal_voyage_count} voyages through Cape Cod Canal in Jan 2014.")
print(f"{Canal_MMSI_count} different ships made those voyages.")
```

_Which ships violated speed restrictions?_

```python
# Identify large, speeding vessels that passed through Cape Cod Canal
IllegalVessels = []

for index, row in Canal.iterrows():
    speed = row['SOG']
    length = row['Length']
    MMSI = row['MMSI']
    Vtype = row['VesselType.y']
    if speed > 10 and length >= 40: # speed limit is 10 knots for vessels over 40'
        IllegalVessels.append((MMSI,Vtype))
        print("Vessel:", MMSI, "Type:", Vtype)
```

```python
# Visualize illegal vessel paths
for item in IllegalVessels:
    q = CleanData.query(f'MMSI == {item[0]}')
    m = q.to_crs(epsg=3857).plot(figsize=(10,8), 
                                 alpha=.5, 
                                 markersize=5)
    cx.add_basemap(m)
```
