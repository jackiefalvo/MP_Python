```python
import pandas as pd
import altair as alt
```

```python
df = pd.read_csv('../Data/Metadata/CCB Metadata.csv', 
                 sep=',', 
                 encoding='utf-8',
                 parse_dates = ['deployDate','recoveryDate'],
                )

df.head()
```

```python
alt.Chart(df).mark_bar().encode(
    x = 'deployDate:T',
    x2 = 'recoveryDate:T',
    color = 'c_recordOnDays'
)
```
