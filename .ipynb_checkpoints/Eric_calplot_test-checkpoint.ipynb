### Parse dates during read

## Date ranges

Expanding date ranges
https://stackoverflow.com/questions/57525566/expand-rows-by-date-range-having-start-and-end-in-pandas

```python
df
```

```python
pd.date_range('1/1/2020','1/20/2020',freq='D')
```

```python
df['between_days'] = df.apply(lambda row: pd.date_range(row['deployDate'],row['recoveryDate'],freq='D'), axis=1)
```

```python
df
```

```python
df.explode('between_days')
```

```python
df_days = df.explode('between_days').reset_index()
df_days
```

```python
df_days['ones'] = 1
```

```python
days_sum = df_days.groupby('between_days').agg({'ones':'sum'})
days_sum
```

```python
calmap.calendarplot(days_sum['ones'],
                   fig_kws=dict(figsize=(12, 8)));
```

```python
import numpy as np; np.random.seed(sum(map(ord, 'calmap')))
import pandas as pd
import calmap

all_days = pd.date_range('1/15/2014', periods=700, freq='D')
days = np.random.choice(all_days, 500)
events = pd.Series(np.random.randn(len(days)), index=days)
```

```python
calmap.yearplot(events, year=2015)
```
