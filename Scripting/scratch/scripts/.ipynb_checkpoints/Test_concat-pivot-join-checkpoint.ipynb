```python
import altair as alt
import pandas as pd
```

```python
# concat3 sample
concat3_sample = pd.read_csv('../data/concat3_sample/Sheet 2-Table 1.csv',
                              parse_dates = ['between_days'])

concat3_sample
```

```python
# concat_pivot sample (came from concat4 -- condensed acoustic df + whale + zoo)
pivot_sample = pd.read_csv('../data/ConcatPivot_sample/Sheet 2-Table 1.csv',
                              parse_dates = ['between_days'])

pivot_sample
```

`melt pivoted data back to long form`

```python
# melt pivot table (back to long form, bring new overlap column along)
pivot_melt = pd.melt(pivot_sample, id_vars=['between_days','overlap'], 
                     value_vars=['acoustic','whale','zooplankton'])

# don't need 'variable field'
p_melt = pivot_melt[['between_days','value','overlap']]
p_melt

# rename 'value' field to 'DataType'
p_melt = p_melt.rename(columns={'value':'DataType'})

p_melt
```

`join p_melt (long df with overlap column) to concat3`

```python
# join pivot_sample to concat3_sample
join = concat3_sample.merge(p_melt, left_on=['between_days','DataType'],
                           right_on=['between_days','DataType'], how='right')

join
```
